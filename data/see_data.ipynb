```python
import numpy as np

# Load the data
data = np.load("dump.npy")

# Print the shape of the array
print("Shape:", data.shape)

# Print the data type
print("Data type:", data.dtype)

# Set NumPy to print the entire array regardless of size
np.set_printoptions(threshold=np.inf, suppress=True, precision=10, floatmode="fixed")

# Print the first 5 items
print("\nFirst 10 items:")
print(data[130:135])

# Print some statistics
print("\nStatistics:")
print("Min value:", np.min(data))
print("Max value:", np.max(data))
```

```python
# Get max value for each column (axis=0 means operate along columns)
column_maxes = np.max(data, axis=1)  # This will be a 1D array of length 512

print(len(column_maxes))

# Calculate mean and std of these maximum values
mean_of_max = np.mean(column_maxes)
std_of_max = np.std(column_maxes)

print(f"Mean of column max values: {mean_of_max}")
print(f"Std dev of column max values: {std_of_max}")

# Optional: print a histogram or distribution of the max values
print("\nFirst 100 column max values:")
print(column_maxes[:100])
```
